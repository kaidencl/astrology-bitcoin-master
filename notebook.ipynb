```python
# 1 for Asc and 10 for MC so I can drop them
```

```python
for column in astro_df.columns.tolist():
    if 't10_' in column:
        astro_df = astro_df.drop(columns=[column])
for column in astro_df.columns.tolist():
    if 't1_' in column:
        astro_df = astro_df.drop(columns=[column]) 
```

```python
for column in astro_df.columns.tolist():
    if '1_' in column:
        astro_df = astro_df.drop(columns=[column])
for column in astro_df.columns.tolist():
    if '_10' in column:
        astro_df = astro_df.drop(columns=[column])  
for column in astro_df.columns.tolist():
    if '_1' in column:
        astro_df = astro_df.drop(columns=[column]) 
```

```python
# scores of planets in bitcoin's birth chart:
# sun: -4, moon: 4, mercury: 9, venus: 11, mars: 11, jupiter: -10, saturn: 2, uranus: -3, neptune: 6, pluto: -4
```

```python
for planet in ['sun', 'moon', 'mercury', 'venus', 'mars', 'jupiter', 'saturn', 'uranus', 'neptune', 'pluto']:
    for column in astro_df.columns.tolist():
        if f't{planet}_' in column:
            bitcoin_planet = str(column).split('_')[2]
            bitcoin_dict = {'sun': -4, 'moon': 4, 'mercury': 9, 'venus': 11, 'mars': 11, 'jupiter': -10, 'saturn': 2, 'uranus': -3, 'neptune': 6, 'pluto': -4}
            astro_df[f'd_{column}'] = bitcoin_dict[bitcoin_planet] * astro_df[f'd_{other_planet}'] * astro_df[column]
        elif f'{planet}_conjunction' in column or f'{planet}_m' in column or f'{planet}_opposition' in column or f'{planet}_sextile' in column or f'{planet}_square' in column or f'{planet}_trine' in column:
            other_planet = str(column).split('_')[2]
            astro_df[f'd_{column}'] = astro_df[f'd_{planet}'] * astro_df[f'd_{other_planet}'] * astro_df[column]
```

### VAR

```python
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
final_features = ['Volume'] + ['Close'] + ['d_moon_conjunction_mercury','d_tsun_trine_saturn',
'd_tsun_conjunction_uranus','d_sun_conjunction_neptune','d_mercury_m_mars','d_tvenus_conjunction_mars',
'd_tvenus_trine_jupiter','d_tvenus_trine_saturn','d_tvenus_trine_sun','d_tjupiter_conjunction_neptune',
'd_tmars_opposition_sun','d_mars_sextile_uranus','d_mars_opposition_saturn','jupiter_conjunction_saturn',
'jupiter_square_uranus','mercury_conjunction_jupiter','moon_square_jupiter','moon_trine_saturn',
'saturn_cap','sun_conjunction_neptune','sun_square_moon','tjupiter_conjunction_venus','tjupiter_sextile_mars',
'tmars_trine_mercury','tmercury_opposition_saturn','tsun_conjunction_neptune','tsun_conjunction_sun',
'turanus_sextile_uranus','turanus_trine_saturn','tvenus_sextile_saturn','venus_conjunction_neptune',
'venus_conjunction_uranus','venus_m_uranus','venus_sextile_neptune',] 
```

```python
var_data = final[final_features].diff(1).dropna()
mod = VAR(endog=var_data, dates=pd.date_range('2015/10/8','2020/7/31',freq='D'))
```

```python
lag_order = mod.select_order()
```

```python
res = mod.fit()
```

```python
for item in final_features:
    print(res.test_causality('Close',[item], kind='wald').summary())
```
